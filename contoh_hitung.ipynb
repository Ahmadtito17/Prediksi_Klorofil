```python
import pandas as pd
import re
import os

# 1. Load Data
file_path = "./Dataset/HH22.xlsx"
df = pd.read_excel(file_path, header=None)
df
```

```python
# 2. Ekstrak R, G, B per baris
def extract_rgb(row):
    match = re.findall(r'R=\s*(\d+(?:\.\d+)?)\s*G=\s*(\d+(?:\.\d+)?)\s*B=\s*(\d+(?:\.\d+)?)', row)
    if match:
        r, g, b = map(float, match[0])
        return pd.Series([r, g, b], index=['R', 'G', 'B'])
    else:
        return pd.Series([None, None, None], index=['R', 'G', 'B'])

# Terapkan ke kolom pertama
rgb_df = df.iloc[:, 0].apply(extract_rgb)

# Hapus baris kosong 
rgb_df.dropna(inplace=True)

# 3. Hitung ExG untuk setiap baris
rgb_df["Excess Green (ExG)"] = (2 * rgb_df["G"]) - rgb_df["R"] - rgb_df["B"]

# # 4. Konversi ExG ke persentase
# exg_min = rgb_df["Excess Green (ExG)"].min()
# exg_max = rgb_df["Excess Green (ExG)"].max()
# rgb_df["ExG (%)"] = ((rgb_df["Excess Green (ExG)"] - exg_min) / (exg_max - exg_min)) * 100

# 5. Simpan ke Excel
file_name = os.path.splitext(os.path.basename(file_path))[0]
output_folder = "./Contoh/"
os.makedirs(output_folder, exist_ok=True)  
output_file = f"Hasil_RGB_ExG_{file_name}.xlsx"
output_path = os.path.join(output_folder, output_file)

rgb_df.to_excel(output_path, index=False)

print(f"Hasil disimpan ke: {output_path}")
print(rgb_df.head())
```
